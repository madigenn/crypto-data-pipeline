```python
df = spark.table("workspace.default.raw_crypto_data_20260716_161210")
display(df)
```

```python
from pyspark.sql.functions import col, when

df_clean = df.select(
    "id",
    "symbol",
    "name",
    "current_price",
    "market_cap",
    "market_cap_rank",
    "total_volume",
    "price_change_percentage_24h",
    "high_24h",
    "low_24h"
)

display(df_clean)
```

```python
df_final = df_clean.withColumn(
    "volatility_flag",
    when(col("price_change_percentage_24h") > 5, "Alta - Positiva")
    .when(col("price_change_percentage_24h") < -5, "Alta - Negativa")
    .otherwise("Normale")
)

display(df_final)
```

```python
df_final.write.format("delta").mode("overwrite").saveAsTable("workspace.default.crypto_clean")
print("✅ Tabella Delta 'crypto_clean' creata con successo!")
```

```python
%sql
SELECT * FROM workspace.default.crypto_clean
ORDER BY market_cap DESC
LIMIT 10
```

```python
# Simuliamo un aggiornamento: ricarichiamo gli stessi dati con una piccola modifica
df_v2 = df_final.withColumn("volatility_flag", 
    when(col("price_change_percentage_24h") > 3, "Alta - Positiva")
    .when(col("price_change_percentage_24h") < -3, "Alta - Negativa")
    .otherwise("Normale")
)

df_v2.write.format("delta").mode("overwrite").saveAsTable("workspace.default.crypto_clean")
print("✅ Tabella aggiornata - creata Versione 1")
```
